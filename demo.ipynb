```python
import matplotlib.pyplot as plt
import numpy as np

def sigmod(x):
    return 1/(1+np.exp(-x))

m = np.arange(-10,10,.01)
plt.plot(m,sigmod(m), label='sigmod')
plt.plot(m,np.tanh(m), label = 'tahn')
plt.grid(True)
plt.legend()
plt.title('Activation Functions')

plt.savefig('sss2.png', dpi = 300, transparent = True)
```
